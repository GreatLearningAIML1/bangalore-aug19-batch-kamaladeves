- you can get the features using .target method

```python
X = iris.data
y = iris.target
```

### Create train and test data

- set a random_state
- test_size: 0.25

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

- convert class vectors (integers) to binary class matrix
- convert labels
- number of classes: 3

```python
y_train = tf.keras.utils.to_categorical(y_train,num_classes=3,dtype='int64')
y_test = tf.keras.utils.to_categorical(y_test,num_classes=3,dtype='int64')
```

```python
model.compile(loss='categorical_crossentropy',optimizer='SGD',metrics=['accuracy'])
```

## Question 4

## Question 5

### Convert data

```python
X = np.array(X)
y = np.array(y)
X = tf.convert_to_tensor(X, dtype=tf.float32)
y = tf.convert_to_tensor(y, dtype=tf.float32)
```

## Question 6

### Normalize data
- Normalize features

```python
X = tf.math.l2_normalize(X,axis=0)
```

## Question 7

### Define weight and bias
- Initialize weight and bias with tf.zeros
- tf.zeros is an initializer that generates tensors initialized to 0
- Specify the value for shape

```python
W = tf.zeros(shape=(4,1))
b = tf.zeros((1))
```

## Question 8

### Get prediction
- Define a function to get prediction
- Approach: prediction = (X * W) + b; here is X is features

```python
def predictor(X,W,b):
  y_pred = tf.add(tf.matmul(X,W),b)
  return y_pred
```

### Calculate loss
- Calculate loss using predictions
- Define a function to calculate loss
- We are calculating mean squared error

```python
def loss(y,y_pred):
  return tf.reduce_mean(tf.square(y-y_pred))
```

## Question 9

### Define a function to train the model
1.   Record all the mathematical steps to calculate Loss
2.   Calculate Gradients of Loss w.r.t weights and bias
3.   Update Weights and Bias based on gradients and learning rate to minimize loss

```python
def train(X,y,W,b,learning_rate=0.01):
  with tf.GradientTape() as tape:
    tape.watch([W,b])
    curr_prediction = predictor(X, W, b)
    curr_loss = loss(y, curr_prediction)

  #evalute the gradient with the respect to the paramters
  dW, db = tape.gradient(curr_loss, [W, b])

  W = W - learning_rate * dW
  b = b - learning_rate * db

  return W,b
```

## Question 10

### Train the model for 100 epochs 
- Observe the training loss at every iteration

```python
for i in range(100):
  W, b = train(X,y,W,b)

  if i % 10 == 0:
    print("Step {:03d}: Loss {:.6f}: ".format(i, loss(y, predictor(X,W,b)).numpy()))

print(f'W : {W.numpy()} , b  = {b.numpy()} ')
```
