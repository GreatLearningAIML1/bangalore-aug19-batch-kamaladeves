```python
#We drop the name column from the original data frame.
df_pk = df.drop(['name'], axis = 1)
```

```python
corr = df_pk.corr()
plt.figure(figsize = (12,12))
sns.heatmap(corr,linewidths=.5)
```

```python
df_pk1 = df_pk.drop(['status'], axis=1)
```

```python
from sklearn import model_selection
from sklearn.model_selection import train_test_split
X = df_pk1
y = pd.DataFrame(df_pk['status'])
test_size = 0.3
seed = 8
X_train,X_test,y_train,y_test = model_selection.train_test_split(X,y,test_size = test_size,random_state=seed)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy')
model.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_predict = model.predict(X_test)
print (metrics.confusion_matrix(y_test,y_predict))
```

```python
 model.score(X_test,y_test)
##Accuracy of test data
```

```python
model.score(X_train,y_train)
##Accuracy of train data
```

```python
print(metrics.classification_report(y_test,y_predict))
```

```python
#We observe that the decision tree model we built for the data set is a typical overfit.(training accuracy =1)
#The model is able to identify all the parkinson's affected people correctly
#But the model has misidentified 6 people in the training set as unhealthy while they are actually healthy.
#The recall value for "1"(parkinson's affected people) is high (1); where as for "0" is just 0.5
#The model is only predicting 50% of the healthy people correctly. 
#But the overall acuuracy score of the model is good
```

```python
# To prune the tree we can use regularisation parameters such as max_depth(to control the max length between a root node and terminal node)
# and min_smaple_leaf(to control the number of leaves before further split)
max_depth_range = list(range(1,10))
accuracy = []
for depth in max_depth_range:
    model = DecisionTreeClassifier(criterion = 'entropy',max_depth = depth, min_samples_leaf = 5, random_state = 8)
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    accuracy.append(score)
```

```python
accuracy
```

```python
plt.plot(max_depth_range,accuracy)
plt.xlabel('max_depth_range')
plt.ylabel('accuracy')
plt.show()
```

```python
#From the above graph we could infer that the accuracy range neither improves nor drops beyond 3 in max_depth_range
#Maximum information gain happens at max_depth_range(3) itself.
#Thus accuracy reduces if we reduce the max_depth below 3 but accuracy remains the same no matter how many times we increase the depth beyond 3.
```

```python
#We fix the max_depth at 3 from the above analysis and try to adjust the min_samples_leaf for a range of values to see if the accuracy improves further
min_smaples_leaf_range = list(range(2,16))
accuracy = []
for leaves in min_smaples_leaf_range:
    model = DecisionTreeClassifier(criterion = 'entropy',max_depth = 3, min_samples_leaf = leaves, random_state = 8)
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    accuracy.append(score)
```

```python
accuracy
```

```python
plt.plot(min_smaples_leaf_range,accuracy)
plt.xlabel('number of leaves')
plt.ylabel('accuracy')
plt.show()
```

```python
#Thus from the above graph we learn that the accuracy of the model decreases beyond 8 leaf nodes.
#As the leaf nodes are generally the result of reduction in entropy,here we could infer that the maximum reduction in entropy has already taken place with minimum depth(3) and minimum number of leaves(2) itself.
#Hence further increase in depth will not yield better accuracy at the same time further increase in number of leaves may lead to reduction in accuracy and model performance.
#Thus regularisation helps us in ensuring if the model is performing at its best.(In this case;as the data set is small,increasing the depth or increasing the leaves did'nt increase the accuracy but we got insight about the best fit of the same(optimal number of depth and leaf nodes) for the model)
#In cases of large data sets regularization might help in pruning the tree and increasing the overall model accuracy and performance
#Thus in case of large data sets we could prevent the overtfitting or under fitting of the tree by using regularisation methods
```

```python
#We build a RandomForest Classifier for the same data set and search for the best hyper parameter using random search.
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
n_estimators = [int(x) for x in np.linspace(start = 2, stop = 20, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(2, 16, num = 10)]
max_depth.append(None)
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
rf = RandomForestClassifier()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_jobs=2,n_estimators=16,criterion="entropy",random_state=8,max_depth=9, min_samples_leaf=1)
model.fit(X_train,y_train)
```

```python
from sklearn import metrics
y_predict = model.predict(X_test)
print (metrics.confusion_matrix(y_test,y_predict))
```

```python
model.score(X_test,y_test)
```
