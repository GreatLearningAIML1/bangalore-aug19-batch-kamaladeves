```python
#Based on the above observation we are taking 5000 as the threshold for rating_counts and
#...4 as the threshold for ratings. If on an average if 5000 people give a rating of 4 and above for 
#..a product, we are considering to be popular among the users.
```

```python
Popular_Products.loc[(Popular_Products['Ratings']>=4) & (Popular_Products['rating_counts']>5000) ]
```

```python
Popular_Products1 = Popular_Products.sort_values(['rating_counts','Ratings'],ascending=[False,False])
```

```python
Popular_Products1.head(5)
```

```python
min_votes = 5000
```

```python
mean_rating_dataset = np.mean(Popular_Products1.Ratings.values)
```

```python
def weighted_rating(Popular_Products1, min_votes, mean_rating_dataset):
    weighted_rating = list()
    V = Popular_Products1.rating_counts.values
    R = Popular_Products1.Ratings.values
    for v, r in zip(V, R):
        if v < 5000:
            weighted_rating.append(0)
        else:
            weighted_rating.append((v/(v+min_votes))*r + (min_votes/(v+min_votes))*mean_rating_dataset)
    return(weighted_rating)
```

```python
Popular_Products1['weighted_rating'] = weighted_rating(Popular_Products1, min_votes, mean_rating_dataset)
```

```python
Popular_Products1.sort_values(['weighted_rating'],ascending = False).head(10)
```

```python
#Thus we have built a popularity based recommended system. Now we can build a user based collaborative filtering model.
```

```python
df2 = df1[df1.groupby('UserId')['UserId'].transform('size') > 100]
```

```python
df2.head(5)
```

```python
from collections import defaultdict
from surprise import SVD
from surprise import Dataset
from surprise import Dataset
from surprise import Reader
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df2[['UserId', 'ProductId', 'Ratings']], reader)

```

```python
from surprise import accuracy
from surprise.model_selection import train_test_split
trainset, testset = train_test_split(data, test_size=0.30,random_state=8)
```

```python
algo = SVD()
algo.fit(trainset)

```

```python
predictions = algo.test(testset)
```

```python
predictions
```

```python
def get_top_n(predictions, n=5):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))
        for uid, user_ratings in top_n.items():
            user_ratings.sort(key=lambda x: x[1], reverse=True)
            top_n[uid] = user_ratings[:n]
    return top_n
            
        
```

```python
top_n = get_top_n(predictions, n=5)
```

```python
top_n
```

```python
for uid, user_ratings in top_n.items():
    print(uid, [iid for (iid, _) in user_ratings])
```
